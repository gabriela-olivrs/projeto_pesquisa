###### Leitura

```python
data_raw = pd.read_csv("~/Documentos/tcc/Desenvolvimento_TCC2/data/data_raw/formulario.csv")
data_raw
```

```python
data_refined = data_raw
```

```python
# Filtrar o DataFrame para remover profissões que não estão envolvidas com TI
profissoes_a_remover = ['Técnico de controle ferroviário', 'Engenheiro Mecânico', 'Consultor de negócios jr com python']
data_refined= data_refined[~data_refined['Cargo Atual / Senioridade:'].isin(profissoes_a_remover)]
```

```python
#Removendo pessoas que nunca trabalhavam homeoffice ou hibrido
data_refined = data_refined.loc[~(data_refined['Você trabalha (ou já trabalhou) no modelo remoto ou híbrido?'].str.contains('não', case=False, na=False))]
```

```python
#Removendo colunas desnecessárias para analise e renomeando as necessárias
colunas_para_remover = ['Carimbo de data/hora', 'Telefone para contato (se deseja participar do sorteio)', 'Você trabalha (ou já trabalhou) no modelo remoto ou híbrido?']

novos_nomes = ['sexo',
'estado_civil',
'idade',
'possui_filhos', 
'idade_media_filhos', 
'cargo_atual/senioridade', 
'home_office/hibrido',
'avaliação_ambiente_trabalho', 
'modelo_ideal', 
'modelo_pre_pandemia',
'modelo_pos_pandemia', 
'satisfacao_vida_pessoal',
'satisfacao_cuidado_filhos',
'satisfacao_tarefas_domesticas',
'satisfacao_estudos', 
'satisfacao_dev_carreira',
'sobrecarga_vida_pessoal',
'sobrecarga_cuidado_filhos',
'sobrecarga_tarefas_domesticas',
'sobrecarga_estudos', 
'sobrecarga_dev_carreira',
'frustracao_vida_pessoal',
'frustracao_cuidado_filhos',
'frustracao_tarefas_domesticas',
'frustracao_estudos', 
'frustracao_dev_carreira',
'inseguranca_vida_pessoal',
'inseguranca_cuidado_filhos',
'inseguranca_tarefas_domesticas',
'inseguranca_estudos', 
'inseguranca_dev_carreira']


# Use o método set_axis para renomear as colunas
data_refined = data_refined.drop(colunas_para_remover, axis=1)
data_refined = data_refined.set_axis(novos_nomes, axis=1, copy=False)
```

```python
# Substituindo 'idade_media_filhos' por 'Não se aplica' quando 'possui_filhos' for igual a 'nao'
data_refined.loc[data_refined['possui_filhos'] == 'Não', 'idade_media_filhos'] = 'Não se aplica'

# Substituindo o nivel de sentimentos por 'Não se aplica' quando 'possui_filhos' for igual a 'nao'
data_refined.loc[data_refined['possui_filhos'] == 'Não', 'satisfacao_cuidado_filhos'] = 'Não se aplica'
data_refined.loc[data_refined['possui_filhos'] == 'Não', 'sobrecarga_cuidado_filhos'] = 'Não se aplica'
data_refined.loc[data_refined['possui_filhos'] == 'Não', 'frustracao_cuidado_filhos'] = 'Não se aplica'
data_refined.loc[data_refined['possui_filhos'] == 'Não', 'inseguranca_cuidado_filhos'] = 'Não se aplica'
```

```python
# Criando dicionários de mapeamento de substituição

#Estado Civil
substituicoes_est_civil = {'Divorciada': 'Divorciado(a)', 'Divorciado': 'Divorciado(a)', 'Namorando (morando junto)' : 'Casado(a)/União Estável', 'Morando junto':'Casado(a)/União Estável', 'Namorido? Mas oficialmente, solteira': 'Solteiro(a)'}
data_refined['estado_civil'] = data_refined['estado_civil'].replace(substituicoes_est_civil)

#Idade
substituicoes_idade = {' anos': ''}
data_refined['idade'] = data_refined['idade'].replace(substituicoes_idade)
```

```python
#Padronização de Cargos

traducao_cargo = {'Data developer':'Desenvolvedor de Dados',
                  'Data Developer':'Desenvolvedor de Dados',
                  'Data Developer Jr.':'Desenvolvedor de Dados',
                  'Senior Software Developer':'Desenvolvedor Software Senior',
                  'Unity Developer / Pleno':'Desenvolvedor Unity',
                  'Junior System Analist':'Analista de Sistemas Júnior',
                  'Analista Data Analytics Jr':'Analista de Dados Jr',
                  'Data Engenieer':'Engenheiro de Dados',
                  'Data Scientist':'Cientista de Dados',
                 'Sênior Analyst':'Analista Sênior',
                 'Sênior Data Scientist':'Cientista de Dados Sênior',
                 'Sênior Data Engineer':'Engenheiro(a) de Dados Sênior',
                  'Data Scientist  Especialista':'Cientista de Dados Especialista'
                 }

data_refined['cargo_atual/senioridade'] = data_refined['cargo_atual/senioridade'].replace(traducao_cargo)
```

```python
padronizacao_cargo = {' jr':'Júnior','jr':'Júnior','JR':'Júnior','/jr':'Júnior','Jr':'Júnior','Junior':'Júnior','junior':'Júnior',
                       'Dev':'Desenvolvedor(a)','developer':'Desenvolvedor(a)','desenvolvedor':'Desenvolvedor(a)','dev':'Desenvolvedor(a)',
                       'desenvolvedora':'Desenvolvedor(a)','Desenvolvedora':'Desenvolvedor(a)','Desenvolver':'Desenvolvedor(a)', 
                       'Desenvolvedor':'Desenvolvedor(a)','Programador':'Desenvolvedor(a)',
                       'Profressora':'Professor(a)','Professor':'Professor(a)',
                       ' pl ':'Pleno',' Pl ':'Pleno',' PL ':'Pleno','Plena':'Pleno','plena':'Pleno','pleno':'Pleno',
                       'Senior':'Sênior','sênior': 'Sênior','senior':'Sênior','Sr':'Sênior','SR':'Sênior',
                       'engenheira':'Engenheiro(a)','engenheiro':'Engenheiro(a)','Engenheira':'engenheira(a)','Engenheiro':'Engenheiro(a)','Engenheira':'Engenheiro(a)',
                       'analista':'Analista','Arquiteto':'Arquiteto(a)','Gestor':'Gestor(a)','Intern':'Estagiário(a)','Estagio':'Estagiário(a)',
                       'Estagiário':'Estagiário(a)','Estagiária':'Estagiário(a)','estagiária':'Estagiário(a)','Estágio':'Estagiário(a)','Estágiario':'Estagiário(a)',
                       'dados':'Dados','fullstack':'Full Stack','Fullstack':'Full Stack','master':'Master','sistemas':'Sistemas',
                       'tecnico':'Técnico(a)','Tecnico':'Técnico(a)','leader':'Leader','Lead':'Leader','Líder': 'Leader',
                       'Front End':'Frontend','front end':'Frontend', 'Diretor':'Diretor(a)',
                       'Técnico':'Tecnico(a)', 'Analista Dados Sênior':'Analista de Dados Sênior','Qa':'QA'}

data_refined['cargo_atual/senioridade'] = data_refined['cargo_atual/senioridade'].replace(padronizacao_cargo, regex=True)
```

```python
remocao_sinais = {'/':' ',
                  ' / ':' ', 
                  '/ ':' ',
                  ' /':' ',
                  ' - ':' ',
                  ' I ' : '',
                  ' II ':'',
                  ' III':'',
                  ' 1':'',
                  ' 2':'',
                  ' 3':'',
                  ' 4':''}

data_refined['cargo_atual/senioridade'] = data_refined['cargo_atual/senioridade'].replace(remocao_sinais, regex=True)
```

```python
padronizacao_idade = {' ano':'',
                  ' anos':'',
                  ' meses':''
                  '11 e 4':'8',
                  '4,5':'4',
                  '13 e 8':'11'}

data_refined['idade_media_filhos'] = data_refined['idade_media_filhos'].replace(padronizacao_idade, regex=True)
```

```python
# Mapear os cargos para os níveis
def classificar_nivel(cargo):
    if 'Sênior' in cargo or \
       'Gerente' in cargo or \
       'Diretor(a)' in cargo  or \
       'Leader' in cargo or \
       'Professor(a)' in cargo or \
       'Master' in cargo or \
       'Executiva' in cargo:
        return 'Avançado'
    
    elif 'Júnior' in cargo or \
         'Estagiário(a)' in cargo or \
         'Auxiliar' in cargo:
        return 'Entrada'
 
    else:
        return 'Intermediário'


# Aplique a função à coluna "Cargo" para criar a nova coluna "Nível"
data_refined['nivel'] = data_refined['cargo_atual/senioridade'].apply(classificar_nivel)

# Exiba o DataFrame resultante
data_refined
```
